```python
from scipy.special import j1
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import ticker, cm
```

```python
N = 100
x = np.linspace(-3.0, 3.0, N)
y = np.linspace(-2.0, 2.0, N)
```

```python
X, Y = np.meshgrid(x, y)
```

```python
rho = np.sqrt(X**2 + Y**2)
z = j1(rho)
```

```python
np.max(z)
```

```python
np.min(z)
```

```python
levels = np.arange(0,1,0.001)
```

```python
fig, ax = plt.subplots()
cs = ax.contourf(X, Y, z, levels, cmap=cm.PuBu_r)
```
